```python
dataset = pd.read_excel('/content/AdventureWorks.xlsx')
```

```python
dataset.tail()
```

```python
# Linhas e Colunas
dataset.shape
```

```python
dataset.dtypes
```

```python
dataset['ID Loja'] = dataset['ID Loja'].astype('object')
dataset['ID Produto'] = dataset['ID Produto'].astype('object')
dataset['ID Cliente'] = dataset['ID Cliente'].astype('object')
```

```python
dataset.dtypes
```

```python
# Receita total

dataset['Valor Venda'].sum()
```

```python
# Custo total
dataset['Custo Total'] = dataset['Custo Unitário'] * dataset['Quantidade']
dataset.head()
```

```python
dataset.head()
```

```python
# Custo Total
dataset['Custo Total'].sum()
```

```python
# Lucro Total
dataset['Lucro Total'] = dataset['Valor Venda'] - dataset['Custo Total']
dataset.head()
```

```python
dataset.head(1)
```

```python
# Lucro Total das vendas
round(dataset['Lucro Total'].sum(),2)
```

```python
# Criando Coluna de Tempo de Entrega
dataset['Tempo de Entrega (dias)'] = (dataset['Data Envio'] - dataset['Data Venda']).dt.days # É preciso transformar a coluna Tempo de Entrega de string para int
dataset.head(1)
```

```python
# Analisando o Tempo médio de entrega por marca
dataset.groupby('Marca')['Tempo de Entrega (dias)'].mean()
```

```python
# Checando Valores NaN
dataset.isnull().sum()
```

```python
# Agrupando o Lucro em relação à ano e marca
dataset.groupby([dataset['Data Venda'].dt.year, 'Marca'])['Lucro Total'].sum()
```
