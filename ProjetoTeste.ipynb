```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
dataset = pd.read_excel('AdventureWorks.xlsx')
# Visualizando as 5 primeiras linhas
dataset.head()
```

```python
dataset.tail()
```

```python
# Linhas e Colunas
dataset.shape
```

```python
dataset.dtypes
```

```python
dataset['ID Loja'] = dataset['ID Loja'].astype('object')
dataset['ID Produto'] = dataset['ID Produto'].astype('object')
dataset['ID Cliente'] = dataset['ID Cliente'].astype('object')
```

```python
dataset.dtypes
```

```python
# Receita total

dataset['Valor Venda'].sum()
```

```python
# Custo total
dataset['Custo Total'] = dataset['Custo Unitário'] * dataset['Quantidade']
dataset.head()
```

```python
# Custo Total Acumulado
dataset['Custo Total'].sum()
```

```python
# Lucro Total
dataset['Lucro Total'] = dataset['Valor Venda'] - dataset['Custo Total']
dataset.head()
```

```python
# Lucro Total das vendas
round(dataset['Lucro Total'].sum(),2)
```

```python
# Criando Coluna de Tempo de Entrega
dataset['Tempo de Entrega (dias)'] = (dataset['Data Envio'] - dataset['Data Venda']).dt.days # É preciso transformar a coluna Tempo de Entrega de string para int
dataset.head(1)
```

```python
# Analisando o Tempo médio de entrega por marca
tempomedio = dataset.groupby('Marca')['Tempo de Entrega (dias)'].mean()
```

```python
# Checando Valores NaN
dataset.isnull().sum()
```

```python
# Agrupando o Lucro em relação à ano e marca
dataset.groupby([dataset['Data Venda'].dt.year, 'Marca'])['Lucro Total'].sum()
```

```python
# Tirando o resultado da forma de notação científica
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
dataset.groupby([dataset['Data Venda'].dt.year, 'Marca'])['Lucro Total'].sum()
```

```python
lucro_ano = dataset.groupby([dataset['Data Venda'].dt.year, 'Marca'])['Lucro Total'].sum().reset_index()
lucro_ano
```

```python
# Qual o total de produtos vendidos, de forma decrescente?
top_produtos = dataset.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True)
```

```python
# Gráfico de Produtos mais vendidos

top_produtos.plot.barh(title = 'Produtos Mais Vendidos')
plt.xlabel('Quantidade Vendida')
plt.ylabel('Produto');
```

```python
# Gráfico de Lucro anual em relação às Marcas
dataset.groupby([dataset['Data Venda'].dt.year,'Marca'])['Lucro Total'].sum().plot.bar(title='Lucro anual (em milhões) x Marcas')
plt.xlabel('Ano Venda, Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='vertical');
```

```python
# Analisando 2009:
# Definindo escopo de análise:
dataset09 = dataset[dataset['Data Venda'].dt.year == 2009]
dataset09
```

```python
# Analisando Relação Lucro x Mês em 2009:

grafico3 = dataset09.groupby(dataset09['Data Venda'].dt.month)['Lucro Total'].sum().plot(title='Lucro x Mês em 2009')
plt.xlabel('Mês')
plt.ylabel('Lucro');
```

```python
# Analisando relação Lucro x Marca em 2009:

dataset09.groupby('Marca')['Lucro Total'].sum().plot.bar(title='Lucro (em milhões) x Marca em 2009')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```

```python
grafico4 = dataset09.groupby('Classe')['Lucro Total'].sum().plot.bar(title='Lucro x Classe em 2009')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```

```python
# Organizando colunas
colunas = dataset.columns.to_list()
colunas = ['Data Venda',
 'Data Envio',
 'Tempo de Entrega (dias)',
 'ID Loja',
 'ID Produto',
 'ID Cliente',
 'No. Venda',
 'Custo Unitário',
  'Custo Total',
 'Preço Unitário',
 'Quantidade',
 'Valor Desconto',
 'Valor Venda',
 'Lucro Total',
 'Produto',
 'Fabricante',
 'Marca',
 'Classe',
 'Cor']

dataset = dataset[colunas]
```
